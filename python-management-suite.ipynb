```python
#Q1 – Online Store Product List
products=["Laptop","Mouse", "Keyboard", "Monitor"]
products.append("Headphones")
print(products)

products.remove("Mouse")
print(products)

total_products=len(products)
print("total number of products:" , total_products)

for product in products:
   print(products)

#why a list is suitable for this scenario?
#because it is easy to add or remove items and maintains order
```

```python
#Q2 – Student Record System
student_record=(1023,"Ali","12-05-2003","Computer Science")
print("student name: " , student_record[1])
print("department: " , student_record[3])

student_record[3]="Software Engineering"
#there will be an error here because tuples cannot be changed


#why tuple is preferred instead of list in this situation?
#A tuple is used instead of a list because student registration information is fixed and should not be modified after it is stored.
```

```python
#Q3 – Event Registration System
registrations={"Ali", "Hamza", "Sara", "Ali", "Ahmed", "Sara", "Zain"}
print(registrations)
#duplicate names (Ali, Sara) are automatically removed because sets only store unique values.

registrations.add("Usman")
print(registrations)

if "hamza" in registrations:
    print("Hamza is registered")
else:
    print("Hamza is not registered")

#why set is useful in this scenario?
#If a student registers multiple times, the set keeps only one entry.
```

```python
#Q4 – Student Marks Management
dictionary= {
    "Ali": 85,
    "Hamza": 90,
    "Sara": 78,
    "Ahmed": 88
}
print("hamza's marks" ,dictionary["Hamza"])

dictionary["Zain"]= 92
print(dictionary)

dictionary["Sara"]= 82
print(dictionary)

for student, mark in dictionary.items():
    print(student, ":", mark)

#why dictionary is best suited for this system?
#A dictionary is best because it stores student names with their marks in key-value pairs, making it easy to access, update, and manage the data.
```

```python
#Q5 – University Course System
course=("AI" ," CS401")
students=["Ali", "Hamza", "Sara", "Ahmed"]
departments={"CS", "SE", "AI"}
marks = {
    "Ali": 85,
    "Hamza": 90,
    "Sara": 78,
    "Ahmed": 88
}
print("Course Information")
print("------------------")
print(f"Course Name: {course[0]}")
print(f"Course Code: {course[1]}")
print("\nEnrolled Students:")
for student in students:
    print(f"- {student}")
print("\nDepartments Offering the Course:")
for dept in departments:
    print(f"- {dept}")
print("\nStudent Marks:")
for student, mark in marks.items():
    print(f"{student}: {mark}")

#why each data structure was used?
#Tuple for course: immutable; List for students: ordered and changeable; Set for departments: unique items; Dictionary for marks: key-value lookup and update.
```
